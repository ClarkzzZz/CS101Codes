**Q0**: Write a function called `multiply_r` that takes two inputs, `x` and `y`, and returns the product of the two inputs. You must solve this recurisvely!

How many times does `multiply_r` get executed?

```python
def multiply_r(x, y):
    if y == 0:
        return 0
    else:
        return x + multiply_r(x, y-1)
    
print(multiply_r(4,3))

    
```

**Q1:** Write a recursive function divide(a, b) that takes two positive integers a (dividend) and b (divisor) as input, and returns the integer quotient obtained by dividing a by b.

You must implement the function using recursion (without using the division / or floor division // operators).
You should simulate division using repeated subtraction.
Ignore the remainder â€” return only the quotient.

```python
def divide(a, b):
    if a < b:
        return 0
    else:
        return divide(a - b, b) + 1
```

**Q2**: Write a function called `has_three` that takes an input, `n`, and returns `True` if 3 is one of the digits in `n`. Otherwise it returns `False`. You must solve this recursively!

```python
def has_three(n):
    if n == 0:
        return False
    elif n%10 == 3:
        return True
    else:
        return has_three(n//10)

print(has_three(232))
```

**Q3:** Write a recursive function that takes a string and return `True` if a string is a palindrome using recursion

<pre>

</pre>

```python
# check if a string is a palindrome using recursion
def palindrome(s):
    if s == "" or len(s) == 1:
        return True
    else:
        if s[0] == s[-1]:
            return palindrome(s[1:-1])
        else:
            return False
```

**Q4:** Write a recursive function that takes a string and return `True` if there are more vowels in the string than consonants.

```python
# check if a string has more vowels than consonants using recursion
def isMoreVowels(S, v=0, c=0):
    if len(S) == 0:
        if v > c:
            return True
        else:
            return False
    else:
        if S[0] in ['a','e','i','o','u']:
            v += 1
        else:
            c += 1
        return isMoreVowels(S[1:], v, c)



print(isMoreVowels('abceo'))
```

```python
def isMoreVowels(s):
    if len(s) == 0:
        return 0, 0
    elif s[0] in ['a','e','i','o','u']:
        return isMoreVowels(s[1:])[0] + 1, isMoreVowels(s[1:])[1]
    else:
        return isMoreVowels(s[1:])[0], isMoreVowels(s[1:])[1] + 1

def checkMoreVowels(s):
    v, c = isMoreVowels(s)
    return v > c

print(checkMoreVowels("abceo"))
```

**Q5:** Write a  recursive method that rearranges an list of integer values so that all the even values appear before all the odd values.
<pre>
    Input: a list of integers
    Return: a re-arranged list of integers
</pre>

```python
def rearrangeEvenOdd(nums):
    if len(nums) == 1:
        return nums
    else:
        if nums[0] % 2 == 0:
            c = []
            c.append(nums[0])
            c = c + rearrangeEvenOdd(nums[1:])
            return c
        else:
            c = []
            c = c + rearrangeEvenOdd(nums[1:])
            c.append(nums[0])
            return c
        

print(rearrangeEvenOdd([num for num in range(10)]))
```

**Q6:** 

Write a recursive function `count_substring_overlap(s, sub)` that counts how many times the substring `sub` appears in the string `s`, including overlapping occurrences. Assume len(s)>1. Some example test cases are as follows: 

count_substring_overlap("aaaaa", "aa") = 4
count_substring_overlap("banana", "ana") = 2
count_substring_overlap("catcowcat", "cat") = 2

```python
def count_substring_overlap(s, sub):
    if s == sub:
        return 1
    elif len(s) <= len(sub):
        return 0
    else:
        length = len(sub)
        if s[ 0 : length ] == sub:
            return count_substring_overlap(s[1:], sub) + 1
        else:
            return count_substring_overlap(s[1:], sub)
```
