## Q1 (10 points)

Given a list of characters letters (`letters`) containing only lowercase letters and a `target` letter, write a function `findLetter(letters, target)` to find the smallest element in the list `letters` that is larger than the given `target`.

Example: findLetter(['d','c','f','k','g'],'a') = 'c'

```python
def findLetter(letters, target):
    """
    :type letters: List[str]
    :type target: str
    :rtype: str
    """
    letters.sort()
    for ch in letters:
        if ch > target:
            return ch
        
```
