## Problem 1 (10 points)

Write a Python function named `sum_of_primes(n)` that calculates and **returns** the sum of all prime numbers between 1 and the given number n (inclusive).

**Note:**: A prime number is an integer greater than 1 that has no positive divisors other than 1 and itself.


**Example usage:**
- sum_of_primes(10) = 17


```python
def sum_of_primes(n):
    """
    Calculate the sum of all prime numbers between 1 and n (inclusive).

    Parameters:
        n (type int): The upper limit (inclusive) of the range to check for prime numbers.

    Returns:
        total (type int): The sum of all prime numbers between 1 and n (inclusive).
    """

    _sum = 0
    
    def _isprime(i):
        res = True
        for j in range(2, i):
            if i % j == 0:
                res = False
        return res

    for i in range(2, n + 1):
        if _isprime(i) == True:
            _sum += i

    return _sum
```

```python
def sum_three_digit_numbers():
    """
    Compute the sum of all three-digit numbers formed using the digits 0, 1, 2, and 3,
    with no repeated digits and no leading zeros.

    Returns:
        answer (type int): The sum of all valid three-digit numbers.
    """
    _sum = 0
    for i in range(1, 4):
        for j in range(0, 4):
            for k in range(0, 4):
                if i == j:
                    continue
                if i == k:
                    continue
                if j == k:
                    continue
                _sum += int( str(i) + str(j) + str(k) )

    return _sum
        
```

## Problem 3 (10 points)

Given a non-negative integer n, write a function **`intToBinary(n)`** that **returns** integer n's binary representation as a **string**. You can use the division-remainder method from week 1 to generate the binary representation.


**Note**: Do not use any built-in functions that directly convert numbers into binary — such as Python’s bin() or format() with binary formatting options. You must perform the conversion manually using arithmetic operations 

    

**Note**:If n is 0, the function should return "0".

**Example usage:**
- intToBinary(13) = "1101"
- intToBinary(0) = "0"


```python
def intToBinary(n):
    res = ''
    if n == 0:
        return str(0)
    while n != 0:
        step = 0
        step = n % 2
        n = n // 2
        res = str(step) + res
    return res
```

## Problem 4 (10 points)
Now, write a new function **`decToBin_fraction(aFraction)`** that converts and **returns** the fractional part (i.e., numbers less than 1, such as 0.25, 0.625, etc.) of a positive decimal number to its binary representation as a **string**. 

Some decimal fractions cannot be represented exactly in binary (for example, 0.1 in decimal is 0.0001100110011… infinitely repeating in binary). To prevent an infinite loop when converting such fractions, the loop is limited to 32 iterations (bits). This means the function will produce up to 32 binary digits after the decimal point, giving a good balance between accuracy and computational efficiency 



**Note**: Do not use any built-in functions that directly convert numbers into binary — such as Python’s bin() or format() with binary formatting options. You must perform the conversion manually using arithmetic operations.

**Example usage:**
- decToBin_fraction(0.25) = "0.01"
- intToBinary(0.1) = "0.00011001100110011001100110011001"

```python
def decToBin_fraction(aFraction):
    binary = "0."
    count = 0  
    while aFraction != 0:
        aFraction *= 2
        if aFraction >= 1:
            binary = binary + '1'
            aFraction -= 1
        else:
            binary = binary + '0'
        count += 1
        if count == 32:
            break
        
    return binary
```

Finally, use `intToBinary(n)` from Problem 3 and `decToBin_fraction(aFraction)` from Problem 4 together to write another function **`decToBinary(num)`**  that converts and **returns** any positive decimal number (e.g., 29.5, 12.375) `num` into its full binary equivalent as a **string**. 

**Example usage:**
- decToBinary(0) == "0"
- decToBinary(5) == "101"
- decToBinary(0.5) == "0.1"
- decToBinary(0.625) == "0.101"
- decToBinary(29.5) == "11101.1"
- decToBinary(12.375) == "1100.011"
- decToBinary(1.0) == "1"
- decToBinary(15.0) == "1111"


```python
def decToBinary(num):
    n = str(num)
    if n.find('.') == -1:
        integer = int(n)
        fraction = 0.0
    else:
        ls = n.split('.')
        integer = int( ls[0] )
        fraction = float( '0.' + ls[1] )

    res1 = ''
    while integer != 0:
        step = 0
        step = integer % 2
        integer = integer // 2
        res1 = str(step) + res1

    res2 = ''
    count = 0
    while fraction != 0:
        fraction *= 2
        if fraction >= 1:
            res2 = res2 + '1'
            fraction -= 1
        else:
            res2 = res2 + '0'
        count += 1
        if count == 32:
            break
    
    if res1 == '':
        res1 = '0'
    if res2 == '':
        return res1
    else:
        return res1 + '.' + res2
```
