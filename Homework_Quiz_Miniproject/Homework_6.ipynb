### Part A: Generate Encoding Keys (5 points)

A book cipher is a cipher for which the key is derived from a book. For example, it might map each character in the plain text to the numeric index **as string** of the first occurrence of that character in the book (or on a page of the book). The assumption is that the sender and receiver of the coded message have previously agreed on the book, but an adversary who intercepts the coded message does not know what book or the page of the book was used to code it.

For example: If plain_text were "no is no" and book started with "Once upon a time, in a house in a land far away," the function `gen_code_keys(book, plain_text)` would return {'n': '1', 'o': '7', ' ': '4', 'i': '13', 's': '26'}. Notice, by the way, that o gets mapped to seven rather than zero because o and O are different characters.

```python
def gen_code_keys(book, plain_text):
    mydict = {}
    for char in plain_text:
        if char not in mydict:
            ind =  book.find(char)
            mydict[char] = str(ind)
    return mydict
```

### Part C:  Decode a Message (5 points)
Write a function `decoder(book, cipher_text)` that deciphers the `cipher_text` back to its original text message based on the `book`

For example: If `cipher_text were "1\*7\*4\*13\*26\*4\*1\*7" and book started with "Once upon a time, in a house in a land far away,", then 
original_text = decoder(book, cipher_text)

```python
def decoder(book, cipher_text):
    res = ""
    chars = cipher_text.split('*')  
    for char in chars:
        res = res + book[int(char)]  
    return res
```

Write a function `gen_safe_code_keys(book, plain_text)` that generates encoding keys safely using this method.


```python
def gen_safe_code_keys(book, plain_text):
    mydict = {}
    book = book + plain_text
    for char in plain_text:
        if char not in mydict:
            ind =  book.find(char)
            mydict[char] = str(ind)
    return mydict
```

## Problem 2 (10 pts)

Given a string s, find the first non-repeating character and return its index. If it does not exist, return -1.

**Note:**
- A character is considered non-repeating if it appears exactly once in the string.
- The comparison is case-sensitive, meaning 'A' and 'a' are treated as different characters.

```python
def first_nonrepeat_char(s):
    """
    Find the first non-repeating character in a string and return its index.
    If no non-repeating character exists, return -1.

    :param s: Input string
    :return: Index of the first non-repeating character or -1
    """
    mydict = {}
    for char in s:
        mydict[char] = mydict.get(char, 0 ) + 1
    for i, char in enumerate(s):
        if mydict[char] == 1:
            return i
    return -1
```

## Problem 3 (10 pts)
Write a function that takes a string and return a dictionary. Your function will need to return the dictionary whose keys are the unique **letters** and the values are the frequency of the **letters** from the string.

**Notes:**
- The function must ignore case (treat uppercase and lowercase letters as the same).
- Non-alphabetic characters such as digits, spaces, and punctuation must not be counted.
- The keys in the dictionary should be lowercase letters only.

Example: 
Sample input string : 'W3resource!'
return: {'w': 1, 'r': 2, 'e': 2, 's': 1, 'o': 1, 'u': 1, 'c': 1}

```python
import string
def count_letters_frequency(input_string):
    input_string = input_string.lower()
    table = '0123456789 ' + string.punctuation
    input_string = input_string.translate(str.maketrans("","",table))
    letter_count = {}
    for char in input_string:
        letter_count[char] = letter_count.get(char, 0) + 1
    return letter_count
```

## Problem 5 (10 points)
Given an array of strings strs, group the anagrams together. You can return the answer in any order.


- Example 1:

    - Input: strs = ["eat","tea","tan","ate","nat","bat"]

    - Output: [['eat', 'tea', 'ate'], ['tan', 'nat'], ['bat']] or any other correct variation of this 

    - Explanation: There is no string in strs that can be rearranged to form "bat". The strings "nat" and "tan" are anagrams as they can be rearranged to form each other. The strings "ate", "eat", and "tea" are anagrams as they can be rearranged to form each other.


- Example 2:

    - Input: strs = [""]

    - Output: [[""]]


- Example 3:

    - Input: strs = ["a"]

    - Output: [["a"]]

 

```python
def group_anagrams(strs):
    anagram_groups = {}
    
    for s in strs:
        key = ''.join(sorted(s))
        
        # Add to the appropriate group
        if key not in anagram_groups:
            anagram_groups[key] = []
        anagram_groups[key].append(s)
    
    return list(anagram_groups.values())
```

## Problem 6 (10 points)

In the file questions.txt, there are 9095 questions from the Stanford Question Answering Dataset, whose question words are in the list ['what', 'who', 'when', 'where', 'how', 'how_many', 'how_much', 'how_old'].

Write a function named count_question_words() that reads the file and computes the frequency of each question word. The function should return a dictionary where:

    The keys are the question words.
    The values are the frequencies (type int) of those question words .

For instance, count_question_words() might return {'when': 736,
                                 'who': 984,
                                 'what': 5813,
                                 'where': 429,
                                 'how_many': 522,
                                 'how': 459,
                                 'how_much': 141,
                                 'how_old': 11}

Important: The keys in the dictionary must appear in the same order as the first occurrence of each question word in the file.

Note: For questions like “How old are you?”, treat the question word as "how_old" rather than "how".

```python
def count_question_words():
    #Do NOT change or delete the following 3 lines of code:
    question_words = ['what', 'who', 'when', 'where', 'how', 'how_many', 'how_much', 'how_old']
    counts = {}
    handle = open('questions.txt', 'r', encoding='UTF-8')

    first_seen_order = []
    
    for line in handle:
        words = line.lower().split()
        
        if words[0] != "how":
            qword = words[0]
        else:
            if len(words) > 1 and words[1] == "many":
                qword = "how_many"
            elif len(words) > 1 and words[1] == "much":
                qword = "how_much"
            elif len(words) > 1 and words[1] == "old":
                qword = "how_old"
            else:
                qword = "how"

        if qword in question_words:
            counts[qword] = counts.get(qword, 0) + 1
            if qword not in first_seen_order:
                first_seen_order.append(qword)

    result = {q: counts[q] for q in first_seen_order}

    return result
```
