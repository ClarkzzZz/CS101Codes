**Note**: The factorial of a non-negative integer n (written as n!) is the product of all positive integers less than or equal to n.

Formally: n! = n $\times (n - 1) \times (n - 2) \times \dots \times 3 \times 2 \times 1$

```python
def myfactorial(num):
    fact = 1
    
    for i in range(1, num + 1):
        fact = fact * i
        
    return fact
```
