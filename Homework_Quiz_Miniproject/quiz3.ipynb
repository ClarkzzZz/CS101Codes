## Q1 (10 points)

Given a string `s` consists of alphabets and empty space characters `' '`, return the length of the penultimate word in the string.

If the penultimate word does not exist, return `-1`.

```python
def myString(s):
    
    words = s.split()
    
    if len(words) < 2:
        return -1
        
    return len(words[-2])
```
