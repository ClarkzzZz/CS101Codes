## Problem 1

Write a function that takes a list and convert the dictionary into a list of lists and return it.

example: Original Dictionary:
{1: 'red', 2: 'green', 3: 'black', 4: 'white', 5: 'black'}

Convert the said dictionary into a list of lists:
[[1, 'red'], [2, 'green'], [3, 'black'], [4, 'white'], [5, 'black']]

```python
def dict_to_list_of_lists(input_dict):
    res = []
    for keys in input_dict.keys():
        c = []
        c.append(keys)
        c.append(input_dict[keys])
        res.append(c)
    return res       
```

```python
assert dict_to_list_of_lists({1:'red',2:'green',3:'black',4:'white',5:'black'})==[[1, 'red'], [2, 'green'], [3, 'black'], [4, 'white'], [5, 'black']]
# hidden tests below
```
