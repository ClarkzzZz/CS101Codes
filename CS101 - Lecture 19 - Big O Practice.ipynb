1) what is the O(?) of the following function?

```python
def intToStr(i):
    digits = '0123456789'
    if i == 0:
        return '0'
    result = ''
    while i > 0:  # Modulus
        result = digits[i % 10] + result
        i = i // 10  # Floor division
    return result
```

2) what does this function do and what is the O(?) of this function?

```python
def intersect(L1, L2):
    tmp = []
    for e1 in L1:
        for e2 in L2:
            if e1 == e2:
                tmp.append(e1)

    res = []
    for e in tmp:
        if not (e in res):
            res.append(e)
    return res
```

4. Using big-Oh, what is the runtime complexity of the call (bleem(N)?

```python
def bleem(n):
    total = 0
    for k in range(n):
        for j in range(0, k, 1):
            total += 1
        for j in range(0, k, 2):
            total += 1
        for j in range(0, k, 3):
            total += 1
    return total
```

5. Using Big-Oh, what are the complexities of the runtime of the call bar(N) and bar(bar(N)), respectively?

```python
def bar(n):
    total = 0
    for k in range(0, n, 2):   
        total += k
    return total
```
